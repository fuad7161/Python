```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def delay(n, k):
    n1 = []
    for i in range(0,len(n)):
        n1.append(n[i]+k)
    return n1
```

```python
def advence(n, k):
    n1 = []
    for i in range(0,len(n)):
        n1.append(n[i]-k)
    return n1
```

```python
def sceling(n , k):
    n1 = []
    for i in range(0,len(n)):
        n1.append(n[i]*k)
    return n1
```

```python
def multiply(a,b):
    n1 = []
    for i in range(len(a)):
        n1.append(a[i]*b[i])
    return n1
```
