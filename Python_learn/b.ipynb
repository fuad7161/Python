```python
def fold(x):
    for i in range(0,len(x)):
        x[i]=(-1)*x[i]
    return x
```

```python
def delay(n, k):
    n1 = []
    for i in range(0,len(n)):
        n1.append(n[i]+k)
    return n1
```

```python
def advence(n, k):
    n1 = []
    for i in range(0,len(n)):
        n1.append(n[i]-k)
    return n1
```
